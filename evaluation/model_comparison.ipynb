## Model Comparison Analysis

## Confusion Matrix Analysis

The confusion matrix indicates that the models correctly classify most samples.
Misclassifications mainly occur between visually similar celebrities, which is
expected in face recognition tasks. This suggests that errors are caused by
similar facial features rather than random predictions.

## Grad-CAM Analysis

Grad-CAM visualizations show that the model focuses on important facial regions
such as the eyes, nose, and mouth. This indicates that the model relies on
meaningful facial features instead of background information, which improves
the explainability of the predictions.

## Challenges and Limitations

## Challenges and Limitations

During the development of this project, several challenges were encountered:

1. **Limited Dataset Size**  
   The dataset contains a relatively small number of images per class, which increased the risk of overfitting, especially for deeper models.

2. **Overfitting in VGG16**  
   VGG16 showed poor generalization performance compared to ResNet50 and EfficientNet. This is mainly due to its large number of parameters and sensitivity to limited training data.

3. **Computational Constraints**  
   Training deep CNN models required significant computational resources. Limited GPU availability increased training time and restricted extensive hyperparameter tuning.

4. **Similar Facial Features Between Classes**  
   Some celebrities share similar facial characteristics, which led to misclassifications as observed in the confusion matrices.

5. **Grad-CAM Interpretation Limitations**  
   While Grad-CAM provided useful visual explanations, the highlighted regions were sometimes broad and not always perfectly localized on facial features.
